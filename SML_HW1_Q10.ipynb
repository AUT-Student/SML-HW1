```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
normal_samples = np.random.normal(0, 1, 10000)
cauchy_samples = np.random.standard_cauchy(10000)
```

```python
normal_samples_average = []
cauchy_samples_average = []

for i in range(10000):
  normal_samples_average.append(np.average(normal_samples[:i+1]))
  cauchy_samples_average.append(np.average(cauchy_samples[:i+1]))
```

```python
fig, ax = plt.subplots(2)
ax[0].scatter(range(1,10001), normal_samples_average, color="green", label="Normal", s=0.1)
ax[1].scatter(range(1,10001), cauchy_samples_average, color="orange", label="Cauchy", s=0.1)
ax[0].set_xlabel("Number of Data")
ax[0].set_ylabel("Average")
ax[1].set_xlabel("Number of Data")
ax[1].set_ylabel("Average")
ax[0].legend()
ax[1].legend()
fig.set_size_inches(15,10)
plt.show()
```
