```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_samples = dict()
for i in [1, 100, 1000, 10000]:
  X_samples[i] = sorted(np.random.normal(0, 1/i, 10000))
min_sample = min(X_samples[1][0], X_samples[100][0], X_samples[1000][0], X_samples[10000][0])
max_sample = max(X_samples[1][0], X_samples[100][0], X_samples[1000][0], X_samples[10000][0])
```

```python
fig, ax = plt.subplots(2,2)

for j, i in enumerate([1, 100, 1000, 10000]):
  ax[j//2][j%2].plot(X_samples[i],
                     np.arange(0, 1, 1/10000),
                     label=f"i={i}",
                     color="purple"
                    )
  
  ax[j//2][j%2].legend()
  ax[j//2][j%2].set_xlabel("X")
  ax[j//2][j%2].set_ylabel("CDF")

fig.set_size_inches(10, 10)
plt.show()
```

```python
epsilon = 0.0001
```

```python
for i in [1, 10, 100, 1000, 10000, 100000]:
  p = np.sum(np.abs(np.random.normal(0, 1/i, 10000)) > epsilon) / 10000
  print(f"i = {i} -> p = {p}")
```
