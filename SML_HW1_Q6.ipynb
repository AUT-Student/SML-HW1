```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.uniform(0, 1, 10*1000*1000)
Y = np.random.uniform(0, 1, 10*1000*1000)
Za = X - Y
Zb = X/Y
```

```python
plt.hist(Za, bins=100, density=True, color="yellow", label="Real")
plt.plot([-1, 0, 1], [0, 1, 0], color="red", ls="-.", label="Estimated")
plt.xlabel("X")
plt.ylabel("f")
plt.title("X - Y")
plt.legend()
plt.show()
```

```python
plt.hist(Zb, bins=100, density=True, color="yellow", label="Real", range=(0, 10))

plt.plot([0, 1], [0.5, 0.5], color="red", ls="-.", label="Estimated")
plt.plot(np.linspace(1, 10, 100), 1/(2*np.linspace(1, 10, 100)**2), color="red", ls="-.")

plt.xlabel("X")
plt.ylabel("f")
plt.title("X / Y")
plt.legend()
plt.show()
```
